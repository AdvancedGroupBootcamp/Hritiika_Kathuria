# Transform

```python
peoples_df.drop(['homeworld', 'species','starships'], axis = 1, inplace = True)
peoples_df.head()
```

```python
peoples_df.rename_axis('id', inplace = True)
```

```python
peoples_df['height'] = peoples_df['height'].apply(lambda x: x.replace(',',''))
peoples_df['mass'] = peoples_df['mass'].apply(lambda x: x.replace(',',''))
```

```python
peoples_df = peoples_df.replace('unknown',np.nan)
```

```python
convert_dict = {
    'name':str,
    'height':float,
    'mass':float,
    'hair_color':str,
    'skin_color':str,
    'eye_color':str,
    'birth_year':str,
    'gender':str,
    'created':str,
    'edited':str,
    'url':str
}

peoples_df = peoples_df.astype(convert_dict)
```

```python
peoples_df.replace(to_replace=[None, 'none', 'n/a', 'unknown'], value=np.nan, inplace=True)
```

```python
peoples_df_films  = peoples_df[[
    'name',
    'films'
]]
```

```python
peoples_df_films = peoples_df_films.explode('films')
peoples_df_films.head()
```

```python
peoples_df_films['films'] = peoples_df_films['films'].apply(lambda s: s.split('/')[5])
```

```python
peoples_df_films
```

```python
peoples_df_vehicles  = peoples_df[[
    'name',
    'vehicles'
]]
```

```python
peoples_df_vehicles = peoples_df_vehicles.explode('vehicles')
peoples_df_vehicles
```

```python
peoples_df_vehicles.dropna(inplace = True)
```

```python
films_df['key'] = films_df['url'].apply(lambda s: s.split('/')[5])
```

```python
peoples_df_vehicles['vehicles'] = peoples_df_vehicles['vehicles'].apply(lambda s: s.split('/')[5] if len(s)>0 else "")
```

```python
peoples_df_vehicles
```

```python
vehicles_df['key'] = vehicles_df['url'].apply(lambda s: s.split('/')[5])
```

```python
films_df.drop(['characters', 'planets','starships','vehicles','species'], axis = 1, inplace = True)
```

```python
vehicles_df.drop(['pilots','films'], axis = 1, inplace = True)
```

```python
peoples_df.drop(['films','vehicles'], axis = 1, inplace = True)
```

# Analysis

```python
con = sqlite3.connect('starwars_database.db')
```

```python
pd.read_sql(
    '''
        SELECT
            *
        FROM
            peoples_films
    
    ''',
    con
)
```

```python
#Characters with skin-color white
pd.read_sql(
    '''
        SELECT
            name,
            gender
        FROM
            peoples
        WHERE
            skin_color = 'white'
        LIMIT 5
    ''',
    con
)
```

```python
#Who appears in most films?
pd.read_sql(
    '''
        SELECT
            name,
            count(*) AS no_films
        FROM
            peoples_films
        GROUP BY
            name
        ORDER BY
            no_films DESC
        LIMIT 5
    ''',
    con
)
```

```python
#Number of vehicles driven by a character
pd.read_sql(
    '''
        SELECT
            name,
            count(*) AS no_vehicles
        FROM
            peoples_vehicles
        GROUP BY
            name
        ORDER BY
            no_vehicles DESC
        
    ''',
    con
)
```

```python
#characters with most film having mass greater than 120 and hair color null

pd.read_sql(
    '''
        SELECT
            f.name,
            count(*) AS no_films
        FROM
            peoples_films f
        INNER JOIN 
            peoples p ON f.name = p.name
        WHERE
            p.hair_color IS NULL
        GROUP BY
            p.name
        HAVING
            p.mass > 120
        ORDER BY
            no_films DESC
        LIMIT 5
    ''',
    con
)
```

```python
#characters and their number of films from films directed by George Lucas whose hair color is not known
pd.read_sql(
    '''
        SELECT
            pf.name,
            count(*) AS no_films
        FROM
            peoples_films pf
        LEFT JOIN 
            films f ON f.key = pf.films
        INNER JOIN
            peoples p ON p.name = pf.name
        WHERE
            f.director = 'George Lucas'
            AND
            p.hair_color IS NULL
        GROUP BY
            pf.name
        ORDER BY
            no_films DESC
        LIMIT 5
    ''',
    con
)
```

```python
#characters from the film 'The Empire Strikes Back' who drive a vehicle made by 'Incom corporation'
pd.read_sql(
    '''
        SELECT
            pv.name,
            count(*) AS no_vehicles
        FROM
            peoples_vehicles pv
        LEFT JOIN 
            vehicles v ON v.key = pv.vehicles
        LEFT JOIN
            peoples_films p ON p.name = pv.name
        INNER JOIN
            films f ON f.key = p.films
        WHERE
            f.key = 1
            AND
            v.manufacturer = 'Incom corporation'
        GROUP BY
            pv.name
        ORDER BY
            no_vehicles DESC
        LIMIT 5
    ''',
    con
)
```

```python
#Making a CTE which creates information about every character, like number of films they star in, number of vehicles driven by them.
#vehicle length >10
pd.read_sql(
    '''
        WITH cte_character_info AS (
            SELECT 
                p.name as character_name,
                f.title as film_name,
                f.director as director_of_the_film,
                v.name as vehicle_driven,
                f.release_date,
                v.length as length_of_vehicle
            FROM
                peoples as p
            INNER JOIN
                peoples_films pf ON pf.name = p.name
            LEFT JOIN
                films f ON f.key = pf.films
            INNER JOIN
                peoples_vehicles pv ON pv.name = p.name
            LEFT JOIN
                vehicles v ON v.key = pv.vehicles
            GROUP BY
                p.name
            )
        SELECT character_name, film_name, release_date, vehicle_driven
        from cte_character_info
        WHERE length_of_vehicle > 10
                
        
    ''',
    con
)
```
