```python
#Le-petit-pont is a pub in the 6eme, with a majestic view of notre dame, and a beautiful terrace which has the heart of everyone who comes for a drink here.

class Employee():
    def __init__(self):
        
        #information about the emloyees
        self.name = None
        self.address = None                 #Address of the emloyees
        self.hours = 0                      #Number of hours they work
        self.job_description = None         #whether they're a bartender, waitress or a manager
        self.salary = 0                     #Amount to be paid to them at the end of the day
        self.is_late = False                #If they come in late, deduct 20e
        self.working_time = None            #Full time or part-time employee
        
    def data_extracting(self, employee_form):
        self.name = employee_form.split('-')[0]
        self.address = employee_form.split('-')[1]
        self.job_description = employee_form.split('-')[2]
        self.hours = int(employee_form.split('-')[3])
        self.is_late = employee_form.split('-')[-1]
        
        
    def data_featuring(self):
        self.street_address = self.address.split(',')[0]
        self.street_name = self.address.split(',')[1]
        self.postcode = self.address.split(',')[-1]
        
    def get_information(self):
        return({
            'Employee name': self.name,
            'Employee postcode' : self.postcode,
            'Employee job': self.job_description,
            'Hours worked' : self.hours,
            'Amount to be paid': self.salary
        })
    
    def find_salary(self):
        if self.job_description == ' waiter ':
            self.salary = self.hours * 10
        elif self.job_description == ' bartender ':
            self.salary = self.hours * 11
        elif self.job_description == ' manager ':
            self.salary = self.hours * 22
        self.cut_salary()
            
#deducting 20e if an employee is late
    def cut_salary(self):
        if self.is_late == ' Yes':
            self.salary = self.salary - 20

#check if an employee is full time            
    def is_full_time(self):
        if self.hours > 20:
            self.woring_time = 'full-time'
        else:
            self.working_time = 'part-time'
        print("Employee {} works {} at Le Petit Pont".format(self.name, self.working_time))
            
            
        
```
