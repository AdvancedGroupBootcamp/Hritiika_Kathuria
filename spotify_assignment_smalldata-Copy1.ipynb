```python
#importing the required libraries
import csv
import json as js
import pandas as pd
```

```python
#loading the file
with open('mpd.json', 'r') as f:
    data = js.load(f)
```

```python
#the list of artists to find information about
list_of_artist = {
    "Zaz",
    "Oumou",
    "Kendrick Lamar",
    "P!nk",
    "Muse",
    "The Killers",
    "Eminem",
    "Taylor Swift",
    "Harry Styles",
    "Lara Fabian",
    "Andrea Bocelli",
    "Laurent Garnier"
}
```

```python
#creating the data frame 
df = pd.json_normalize(data["playlists"])
df
```

```python
dd = pd.json_normalize(df["tracks"])
dd
```

```python
#creating a dict of the artists and saving empty values 

Data_extracted = {}

for artist in list_of_artist:
    Data_extracted[artist] = {
        #"name": artist,
        "no_playlist": 0,
        "duration": 0,
        "tracks": 0,
        "list": []
    }
```

```python
#the empty dict with all the information
Data_extracted
```

```python
#looping through all the playlists and their tracks to get our required data
for playlist in data["playlists"]:
    for track in playlist["tracks"]:
        #print(track)
        if(track['artist_name'] in Data_extracted.keys()):  
            
            Data_extracted[track['artist_name']]['tracks'] += 1            
            Data_extracted[track['artist_name']]['duration'] += track['duration_ms']/60000 #convert from ms
            Data_extracted[track['artist_name']]['list'].append(str(playlist["name"]))
```

```python
#finding the number of playlist an artist was added to
for key in Data_extracted:
    k = len(set(Data_extracted[key]['list']))
    Data_extracted[key]['no_playlist'] = k
```

```python
#final answer
Data_extracted
```

```python
#final_answer = pd.DataFrame(list(Data_extracted.items()), columns = ['Name','Playlist', 'Duration', 'Tracks', 'List'])
final_answer = pd.DataFrame.from_dict(Data_extracted, orient = 'index')
```

```python
del final_answer['list']
```

```python
#data frame with all the answers
final_answer
```

```python
#finding the artists with maximum tracks, playlists and duration played
artist_max_playlist = final_answer[['no_playlist']].idxmax()
artist_max_track = final_answer[['tracks']].idxmax()
artist_max_duration = final_answer[['duration']].idxmax()

max_duration = final_answer['duration'].max()

#converting to hours, mins, and seconds frame
def convert_ms(millis):
    seconds=millis
    seconds = int(seconds)
    minutes=(millis)%60
    minutes = int(minutes)
    hours=(millis/60)%24

    print ("%d:%d:%d" % (hours, minutes, seconds))

```

```python
#displaying the final result
print('Artist with the most playlist features:',  artist_max_playlist)
print('Artist with the most tracks:', artist_max_track)
print('Artist with the longest duration:', artist_max_duration)
convert_ms(max_duration)
```
